```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def run(seed=100):
    x=10
    dt=0.001
    np.random.seed(seed)
    nsteps=10000
    traj=[]
    for istep in range(nsteps):
        x+=-x*dt + np.sqrt(2.0*dt)*np.random.normal()
        traj.append(x)
    return np.array(traj)
```

```python
plt.plot(run())
plt.show()
```

```python
def run(seed=100):
    dt=0.001
    np.random.seed(seed)
    nsteps=10000
    ntrajs=100
    
    trajs=[]
    for i in range(ntrajs):
        x=10
        traj=[]
        for istep in range(nsteps):
            x+=-x*dt + np.sqrt(2.0*dt)*np.random.normal()
            traj.append(x)
        trajs.append(traj)
    return np.array(trajs)
```

```python
trajs=run()
```

```python
for i in range(len(trajs)):
    plt.plot(trajs[i,:])
plt.show()
```

```python
time=0.001*np.array(range(trajs.shape[1]))
plt.plot(time,np.average(trajs,axis=0),label="<x>")
#plt.plot(time,np.average(trajs**2,axis=0),label="<x**2>")
plt.plot(time,np.std(trajs,axis=0),label="stddev")
plt.plot(time,10*np.exp(-time))
plt.plot(time,np.sqrt(1-np.exp(-2*time)))
plt.legend()
```

```python
for n in [10,5,2,1]:
    plt.plot((trajs[:,n:]-trajs[:,:-n]).flatten()[::100],".",label=str(n))
plt.legend()
```

```python
def run(seed=100):
    dt=0.001    
    nsteps=10000
    ntrajs=100
    
    x0=np.linspace(0,10,ntrajs)
    
    trajs=[]
    for i in range(ntrajs):
        np.random.seed(seed)
        x=x0[i]
        traj=[]
        for istep in range(nsteps):
            x+=-x*dt + np.sqrt(2.0*dt)*np.random.normal()
            traj.append(x)
        trajs.append(traj)
    return np.array(trajs)
```

```python
trajs=run()
```

```python
for i in np.array(range(len(trajs)))[::20]:
    plt.plot(trajs[i,:])
plt.show()
```
